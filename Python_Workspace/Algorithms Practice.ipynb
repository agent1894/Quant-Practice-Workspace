## 插入排序

算法复杂度 $\Theta(n^2)$

```python
def insert_sort(array, method=1):
    count = 0
    if method == 1:
        for i in range(1, len(array)):
            temp = array[i]
            for j in range(i, 0, -1):
                if array[j] < array[j-1]:
                    array[j] = array[j-1]
                    array[j-1] = temp
                count += 1
                print(array)
        print('----------')
        print('Operation Times:', count)
        print('Sorted Array:', array)
    else:
        for j in range(1, len(array)):
            for i in range(1, len(array)):
                temp = array[i]
                if array[i] < array[i-1]:
                    array[i] = array[i-1]
                    array[i-1] = temp
                count += 1
                print(array)
        print('----------')
        print('Operation Times: {}'.format(count))
        print('Sorted Array: {}'.format(array))


array1 = list(range(10, 0, -1))
insert_sort(array1, 1)
array2 = list(range(10, 0, -1))
insert_sort(array2, 2)
```
